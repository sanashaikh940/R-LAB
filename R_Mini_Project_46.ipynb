```r
dat <- read.csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv",header=T,sep=";")
dat
```

```r
# 4-A. find out how many values are missing in each of the columns of your data set.
colSums(is.na(dat))
```

```r
# b. select colum with more than 10% values missing for deletation. Create new copy of dataset
# that has one less column call it sample_data_1less.

## Remove columns with more than 10% NA
dataless1 <-dat[, which(colMeans(!is.na(dat)) > 0.1)]
dataless1
```

```r
# d. For attribute having maximum spread delete all tuples that have value less than the mean of the
# attribute selected. Count the number of tuples reamining in the sampled dataset.

m <- mean(dat$fixed.acidity)
m

df2 <- subset(dat, dat$fixed.acidity> m)
summary(df2$fixed.acidity)
print('Rows before row deletion: ')
nrow(dat)
print('Rows before row deletion: ')
nrow(df2)
```

Draw a box plot for each of the two attributes selected above in a single plot

```r
datta <- data.frame(values=c(dat$volatile.acidity,dat$sulphates),group=c(rep("Volatile acidity",length(dat$volatile.acidity)),rep("Sulphates",1599)))
boxplot(values ~ group,datta,col=c("violet","yellow"))
install.packages("ggplot2")
library("ggplot2")
ggplot(datta,aes(x=group,y=values,fill=group))+
geom_boxplot()
```

Write a R function to plot histogram of some specific input attribute for any data set provided

```r
colc <- c("fixed.acidity",	"volatile.acidity",	"citric.acid",	"residual.sugar",	"chlorides",	"free.sulfur.dioxide",	"total.sulfur.dioxide",	"density",	"pH",	"sulphates",	"alcohol","quality") 
usr <- readline("Enter Attribute")
b<-0
for(i in 1:length(colc)){
  if(colc[i]==usr){
    b<-i
    break
  }
  
}
switch(b,hist(dat$fixed.acidity),hist(dat$volatile.acidity),hist(dat$citric.acid),hist(dat$residual.sugar),hist(dat$chlorides),hist(dat$free.sulfur.dioxide),hist(dat$free.density),hist(dat$pH),hist(dat$sulphates),hist(dat$alcohol),hist(dat$quality))
```
